# Posstest 3 Visualisasi Data

## Import Library

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("final_cars_datasets.csv")
```

```python
df.head()
```

## Grapik Pie
Dari grafik pie di bawah dapat dilihat bahwa sistem 2WD(2 Wheel Drive) adalah sistem yang paling banyak digunakan dikalangan mobil-mobil yaitu sebesar 91,5%.
Sedangkan AWD (All Wheel Drive) adalah sistem yang paling sedikit digunakan yaitu sebesar 0,2%

```python
# Menghitung kolum drive ke variabel data
data = df["drive"].value_counts().rename_axis("drive").reset_index(name="id_drive")

plt.figure(figsize=(10,10))
# Menampilkan grafik pie
plt.pie(
    data["id_drive"],
    labels = data["drive"],
    autopct='%.1f%%',
)
# Memberi judul grafik
plt.title("Data Drive System")
# Menampilkan keterangan
plt.legend()
plt.show()
```

## Grafik Bar
Grafik bar di bawah menunjukkan jumlah transmisi yang digunakan mobil-mobil pada dataset.
<ul>
    <li>Automatic : Transmisi yang paling banyak digunakan dengan total > 2000</li>
    <li>Manual    : Transmisi manual digunakan dengan total < 500 </li>
    <li>CVT       : Transmisi yang paling sedikit digunakan dengan total < 100  </li>
</ul>

```python
# Menampilkan boxplot transmisi
sns.countplot(data=df,x="transmission",palette="bright")

# Memberi xlabel,ylabel, dan judul grafik
plt.xlabel("Transmisi Mobil")
plt.ylabel("Total")
plt.title("Total Kategori Transmisi Mobil")
plt.show()
```

```python
plt.figure(figsize=(6,6))
# Menampilkan boxplot transmisi dan harga
sns.boxplot(x=df["transmission"], y=df["price"],orient="v",palette="dark")

# Memberi xlabel,ylabel, dan judul grafik
plt.xlabel("Transmisi")
plt.ylabel("Harga")
plt.title("Boxplot Transmisi dan Harga")
plt.show()
```
